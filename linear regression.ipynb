```python
df.columns = ['named','name','fuel','year','km_driven','owner','location','transmission','car_type','price']
print(df.columns)
```

```python
#convert data types
df.columns = df.columns.str.strip().str.lower()

df['year'] = pd.to_numeric(df['year'], errors='coerce')
df['km_driven'] = pd.to_numeric(df['km_driven'], errors='coerce')
df['price'] = pd.to_numeric(df['price'], errors='coerce')

print(df.head())
```

```python
print(df.dropna(inplace=True))
```

```python
#What this does:

#You are creating a new column called car_age.

# Why?
#Raw data has year (e.g., 2018, 2020)
#But models understand age better than year

# Example:

#year	car_age
#2020	6
#2015	11
# Logic:
#car_age = current_year - manufacturing_year
# Why important?
#Age directly affects price
#Older car → lower price
#Helps model learn better pattern
#Feature enff
```

```python
df = pd.get_dummies(df, drop_first=True)
#ML models cannot understand text
#Example:
#fuel: Petrol, Diesel
#Converted to:
#fuel_Diesel = 0 or 1
```
